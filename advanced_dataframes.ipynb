# Copy the users and roles DataFrames from the examples above.


```python
users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['bob', 'joe', 'sally', 'adam', 'jane', 'mike'],
    'role_id': [1, 2, 3, 3, np.nan, np.nan]
})
users
```

```python
roles = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})
roles
```

# What is the result of using a right join on the DataFrames?


```python
users.merge(roles, left_on = 'role_id', right_on='id', how='right', indicator = True)
```

# What is the result of using an outer join on the DataFrames?


```python
users.merge(roles, left_on = 'role_id', right_on='id', how='outer', indicator = True)
```

# What happens if you drop the foreign keys from the DataFrames and try to merge them?


```python
# you wont be able to join the tables
# we get a key error on 'role_id'
```

# Using the mpg dataset, find out which which manufacturer has the best miles per gallon on average?


```python
# volkswagen has the best mileage at 39.5
mpg.groupby('manufacturer').average_mileage.max()
```

```python
# volkswagen has the best milage at 28.5
mpg.groupby('manufacturer').average_mileage.mean().nlargest(1)
```

```python
# alternative ways to find answer
mpg.groupby('manufacturer').average_mileage.mean().sort_values()
mpg.groupby('manufacturer').average_mileage.mean().nlargest(5)
```
